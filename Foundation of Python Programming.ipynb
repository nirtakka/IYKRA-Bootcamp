#### Convert dict to DataFrame

```python
code = ['US', 'AUS', 'JPN', 'IN', 'RUS', 'MOR', 'EG']
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr = [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]

# Create dictionary my_dict with three key:value pairs: my_dict
my_dict = {'ctrcode':code, 'country':names, 'drives_right':dr, 'cars_per_cap':cpc}
cars = pd.DataFrame(my_dict)

print(type(cars))
print(cars)
```

#### Pandas DF Quick Look

```python
# Dimension
cars.shape # Row dan column pada cars
print(cars.columns) # Nama-nama column pada cars
print(len(cars.columns)) # Jumlah column pada cars
print(len(cars)) # Jumlah row pada cars
```

```python
# Brief View
cars.head(3)
cars.tail(3)
```

```python
# Data type
cars.dtypes # Tipe-tipe column pada cars
cars.info() # Informasi cars
```

```python
# Summary Statistics
cars.describe()
```

```python
# Set DF Column Index
cars_id = cars.set_index("ctrcode")
print(cars_id.shape)
cars_id
```

#### Pandas loc and iloc

Square brackets. There’s fundamental difference between using single and double bracket.
<br>Column access : cars_id['country'], cars_id[['country']]
<br>Row access : cars_id[0:3]
<br>
<br>LOC (label based)
<br>Row access : cars_id.loc[['AUS', 'EG’]]
<br>Row & Column access : cars_id.loc[['RUS', 'MOR'], ['country', 'drives_right']]
<br>
<br>iLOC (integer index based)
<br>Row access : cars_id.iloc[[1, 6]]
<br>Row & Column access : cars_id.iloc[[4,5], [1,2]]

```python
# Single bracket becoming Series
# Column access
cars_id['country']
```

```python
# Double bracket becoming DataFrame
# Column access
cars_id[['country']]
```

```python
# Row access
cars_id[0:3]
```

```python
# LOC
# Row access
cars_id.loc[['AUS', 'EG']]
```

```python
# LOC
# Row & Column access
cars_id.loc[['RUS', 'MOR'], ['country', 'drives_right']]
```

```python
# iLOC
# Row access
cars_id.iloc[[1, 6]]
```

```python
# iLOC
# Row & Column access
cars_id.iloc[[4,5], [1,2]]
```

#### Where Clause and Sort

```python
# Where
print(cars_id[cars_id.drives_right == True])
print(cars_id[(cars_id.cars_per_cap > 100) & (cars_id.country == "Australia") ])
```

```python
# Sort
cars_id.sort_values(by=['cars_per_cap'])
cars_id.sort_values(by=['cars_per_cap'], ascending=False)
```

### Exercise

#### 1. How many rows and columns reflected on dataset file IMDB-database-update.csv?

```python
row = imdb.shape[0]
column = imdb.shape[1]

print("Number of rows reflected on dataset file IMDB-database-update.csv are " + str(row)) 
print("\nNumber of columns reflected on dataset file IMDB-database-update.csv are " + str(column)) 
```

#### 2. What are the top 3 movies with highest nett profit in 2015?

```python
movie_2015 = imdb[imdb.title_year == 2015]

nett_2015 = movie_2015.sort_values(by=['net_USD'], ascending = False).head(3)

top_2015 = nett_2015[["movie_title"]].reset_index(drop=True)

print("Top 3 movies with highest nett profit in 2015:")
top_2015
```

#### 3. What are the top 3 movies with highest gross profit in 2016?

```python
movie_2016 = imdb[imdb.title_year == 2016]

gross_2016 = movie_2016.sort_values(by=['gross_USD'], ascending = False).head(3)

top_2016 = gross_2016[["movie_title"]].reset_index(drop=True)

print("Top 3 movies with highest gross profit in 2016:")
top_2016
```

#### 4. How many non-english movies are in the dataset?

```python
non_eng = imdb[imdb.language != 'English']
print("There are " + str(non_eng.shape[0]) + " non-english movies")
```

#### 5. What are top 3 languages in the data set aside English (based on number of movies)?

```python
top_lang = non_eng["language"].value_counts().head(3)
print("Top 3 languages in the data set aside English:")
top_lang
```

#### 6. What are the top 3 countries with the highest nett profit in all dataset times?

```python
group_country = imdb[["country", "net_USD"]].groupby(by='country').sum().sort_values(by =["net_USD"], ascending = False).head(3)

print("The top 3 countries with the highest nett profit in all dataset times:")
group_country
```

#### 7. What are the average budget of all movies in dataset in 2015?

```python
avg_budget_2015 = round(movie_2015["budget_USD"].mean(), 3)

print("Average budget of all movies in dataset in 2015 is $" + str(avg_budget_2015))
```

#### 8. What are the average budget of all movies in dataset in 2016?

```python
avg_budget_2016 = round(movie_2016["budget_USD"].mean(), 3)

print("Average budget of all movies in dataset in 2016 is $" + str(avg_budget_2016))
```

#### 9. Which one is higher, average budget of all movies in dataset in 2015 or 2016?

```python
if avg_budget_2015 > avg_budget_2016:
    print("Average budget of all movies in dataset in 2015 is higher than in 2016")
else:
    print("Average budget of all movies in dataset in 2016 is higher than in 2015")
```

#### 10. What movie that resulted highest nett profit of all dataset times? How much did the movie make?

```python
nett = imdb.sort_values(by=['net_USD'], ascending = False).head(1)
title = nett[["movie_title"]].iloc[0][0]
profit = round(nett[["net_USD"]].iloc[0][0], 3)

print("Movie that resulted highest nett profit of all dataset times is " + title + " with nett profit $" + str(profit))
```
